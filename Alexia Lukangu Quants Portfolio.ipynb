###### Moving Average Trading Strategy

I will create 2 strategies one with short lookback period and another with long lookback period. Perfect for a systematic trading of one stock. Whenever the two metric cross I am going to take a position, either short or call 

```python
s_lookback = 45
l_lookback = 90
```

```python
df2['s_lookback']=df2['Adj Close'].rolling(window=s_lookback).mean()
df2['l_lookback']=df2['Adj Close'].rolling(window=l_lookback).mean()
```

Overall we can observe that the moving averages and adjusted close price follow the same trend

```python
df2[['Adj Close','s_lookback', 'l_lookback']].plot(title= "NVDA Stock Price Chart", grid=True,
                                                   linewidth=1, figsize=(14,9), fontsize=(14))
```

we define 1 as long/ buyig and -1 as short/ selling , so de define a position whereas if the shortest lookback exced the longest then we buy 


```python
df2['position'] = np.where(df2['s_lookback'] > df2['l_lookback'], 1, -1)
```

```python
df2.dropna(inplace=True)
```

```python
df2
```

We can observe our position over the years. For example based on out strategies the year 2020 has mostly +1 position (buy)

```python
df2[['Adj Close','s_lookback', 'l_lookback', 'position']].plot(secondary_y='position', grid=True, linewidth=1, figsize=(14,9), fontsize=(14))
```

###### Comparison on the Created Strategy and the Buy and Hold Strategy 

We can compare the returns of this strategy to a buy and hold strategy
The comparison shows the bnh startegy starts well .It shows that there was a period in 2019 that the returns were below 50%. The moving averges performed the same during the end 2020 to the whole year of 2021 underperformed year 2022 , and overperformed for the rest of the periods and also during the end of 2019 and most part of 2020.

```python
df2['bnh_returns'] = np.log(df2['Adj Close'] / df2['Adj Close'].shift(1))
df2['strategy_returns'] = df2['bnh_returns'] * df2['position'].shift(1)

df2[['bnh_returns', 'strategy_returns']].cumsum().plot(grid=True, figsize=(12, 8), linewidth=1)
```
